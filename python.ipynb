Displacement Current Density. A magnetic flux density, $\vec{B}= \hat{y}0.1\cdot (\cos{100t})(\cos{5z})$ [T] exists in a linear, isotropic,
homogeneous material characterized by $\epsilon$ and $\mu$. Find the displacement current density in the material if there are no
source charges or current densities in the material.


Densidad de corriente de desplazamiento. Existe una densidad de flujo magnético, $\vec{B}= \hat{y}0.1\cdot (\cos{100t})(\cos{5z})$ [T] en un material lineal, isotrópico y homogéneo caracterizado por $\epsilon$ y $\mu$ Encuentre la densidad de corriente de desplazamiento en el material si no hay cargas de fuente o densidades de corriente en el material. 

###Solución:

$$\vec{B}= \hat{y}0.1\cdot (\cos{100t})(\cos{5z})$$

$$\vec{H}=\frac{\vec{B}}{\mu}$$

$$\nabla\times\vec{H}=\vec{J}+\frac{\partial \vec{D}}{\partial t} $$

$$\vec{J}=0$$

$$\frac{\partial \vec{D}}{\partial t}=\nabla \times \vec{H}$$

$$\nabla \times \vec{H}=\nabla \times \left(\frac{1}{\mu}\hat{y}0.1\cdot (\cos{100t})(\cos{5z})\right)$$

$$\nabla \times \vec{H}=\frac{1}{\mu}(\sin(5z)\cos(100t))\hat{x}$$

```python

import sympy as sym #LIBRERIAS PARA CÁLCULO SIMBOLICO
from sympy.physics.vector import ReferenceFrame
from sympy.physics.vector import curl

#constantes
#@markdown Tiempo de evaluación:
t=1 #@param{type: "number"}
#@markdown Costante $\mu$:
mu = 1 #@param{type: "number"}
#@markdown Costante $\epsilon$:
epsilon = 12.1 #@param {type:"number"}
R = ReferenceFrame('R')

B = 0.1*sym.cos(100*t)*sym.cos(5*R[2])*R.y
H = B/mu
G = curl(H, R) 
G
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

#@markdown ancho y largo de figura:
width= 10 #@param{type:"number"}
height= 10 #@param{type:"number"}
fig = plt.figure()
fig.set_size_inches(width, height)#tamaño de imagen
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

u = 4.16667*np.sin(5*y)*np.cos(100)
v = 0
w = 0

ax.quiver(x, y, z, u, v, w, length=0.1)
plt.title("Densidad de corriente de desplazamiento")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
